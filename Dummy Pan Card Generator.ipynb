```python
from PIL import ImageDraw
from PIL import Image
from PIL import ImageFont
import requests
import ssl
import datetime
import random
import string
from io import BytesIO
import re

non_url_safe = ['"', '#', '$', '%', '&', '+',
                    ',', '/', ':', ';', '=', '?',
                    '@', '[', '\\', ']', '^', '`',
                    '{', '|', '}', '~', "'"]

def slugify(text):
    """
    Turn the text content of a header into a slug for use in an ID
    """
    non_safe = [c for c in text if c in non_url_safe]
    if non_safe:
        for c in non_safe:
            text = text.replace(c, '')
    # Strip leading, trailing and multiple whitespace, convert remaining whitespace to _
    text = u'_'.join(text.split())
    return text

def randomPANNum(stringLength=4):
    """Generate a random string of fixed length """
    letters = string.ascii_uppercase
    return ''.join(random.choice(letters) for i in range(stringLength)) + str(random.randint(1020,8000))

ssl._create_default_https_context = ssl._create_unverified_context


import matplotlib.pyplot as plt

image = Image.open("resources/empty_pan.png")

font = ImageFont.truetype("Verdana", 27)
signature = ImageFont.truetype("Autograf", 46)

def generate_pan(user):
    image = Image.open("resources/empty_pan.png")
    
    ImageDraw.Draw(
        image  # Image
    ).text(
        (34, 140),  # Coordinates
        user['name'].upper(),  # Text
        (0, 0, 0),  # Color
        font=font
    )

    ImageDraw.Draw(
        image  # Image
    ).text(
        (34, 180),  # Coordinates
        user['fname'].upper(),  # Text
        (0, 0, 0),  # Color
        font=font
    )

    ImageDraw.Draw(
        image  # Image
    ).text(
        (34, 240),  # Coordinates
        user['dob'].upper(),  # Text
        (0, 0, 0),  # Color
        font=font
    )

    ImageDraw.Draw(
        image  # Image
    ).text(
        (34, 320),  # Coordinates
        randomPANNum(),  # Text
        (0, 0, 0),  # Color
        font=font
    )
    
    ImageDraw.Draw(
        image  # Image
    ).text(
        (34, 380),  # Coordinates
        user['name'].split(' ')[0],  # Text
        (0, 0, 0),  # Color
        font=signature
    )

    response = requests.get("https://i.pravatar.cc/150")
    photo = Image.open(BytesIO(response.content))

    image.paste(photo, (560, 300))
    rgb_im = image.convert('RGB')
    rgb_im.save('resources/dummy_pans/'+slugify(user['name'])+'.jpg', quality=95)
    return rgb_im

for user in users:
    generate_pan(user)
```
